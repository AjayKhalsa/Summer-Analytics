```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Pokemon.csv', index_col=0,encoding='latin1')
```

```python
df.head()
```

```python
sns.set_style('darkgrid')
sns.lmplot(x='Attack', y='Defense',data=df,
          fit_reg=False,
          hue='Stage')
plt.xlim=(0,None)
plt.ylim=(0,None)
```

```python
sns.boxplot(data=df)
```

```python
stats_df = df.drop(['Total','Stage','Legendary'],axis=1)
sns.boxplot(data=stats_df)
```

```python

sns.violinplot(x='Type 1', y='Attack',data=df)
```

```python
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
```

```python

plt.figure(figsize=(10,6))
sns.violinplot(x='Type 1', y='Attack',data=df,palette=pkmn_type_colors)
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(x='Type 1',y='Attack',data=df,
             palette=pkmn_type_colors)
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='Type 1',y='Attack',data=df,inner=None, palette = pkmn_type_colors)
sns.swarmplot(x='Type 1',y='Attack',data=df,color='k',alpha=0.7)
plt.title('Attack By Type')
```

```python
melted_df =pd.melt(stats_df,
                  id_vars=['Name','Type 1','Type 2'],
                  var_name="Stat")
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(x='Stat',y='value',data=melted_df,hue='Type 1',dodge=True,palette=pkmn_type_colors)

plt.legend(bbox_to_anchor=(1,1),loc=2)
```

```python
plt.figure(figsize=(10,6))

sns.lmplot(x='Attack', y='Defense',data=df,
          fit_reg=False,
          hue='Stage')
sns.kdeplot(df.Attack,df.Defense)

# plt.xlim=(0,None)
# plt.ylim=(0,None)
```

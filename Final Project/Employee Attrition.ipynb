```python
subdf = df.drop(['Id','Attrition','EmployeeNumber','Behaviour'],axis=1)
# Age
# DistanceFromHome
# Education
# EnvironmentSatisfaction
# JobInvolvement
# JobSatisfaction
# MonthlyIncome
# NumCompaniesWorked
# StockOptionLevel
# TotalWorkingYears
# TrainingTimesLastYear
# YearsAtCompany
# YearsInCurrentRole
# YearsSinceLastPromotion
# YearsWithCurrManager
# CommunicationSkill
# JobRole
# BusinessTravel
# EducationField
# Department
# Gender 
# MaritalStatus
# OverTime
```

```python
business_dummies = pd.get_dummies(subdf.BusinessTravel,prefix='Travel').iloc[:,1:]
dept_dummies = pd.get_dummies(subdf.Department,prefix='Dept').iloc[:,1:]
edu_dummies = pd.get_dummies(subdf.EducationField,prefix='Edu').iloc[:,1:]
g_dummies = pd.get_dummies(subdf.Gender,prefix='g').iloc[:,1:]
job_dummies = pd.get_dummies(subdf.JobRole,prefix='j').iloc[:,1:]
mar_dummies = pd.get_dummies(subdf.MaritalStatus,prefix='mar').iloc[:,1:]
over_dummies = pd.get_dummies(subdf.OverTime,prefix='over').iloc[:,1:]

df_with_dummies = pd.concat([subdf,business_dummies,dept_dummies,edu_dummies,g_dummies,job_dummies,mar_dummies,over_dummies],axis='columns')
df_with_dummies.head()
```

```python
df_with_dummies=df_with_dummies.drop(['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','OverTime'],axis=1)
# df_with_dummies.columns
```
